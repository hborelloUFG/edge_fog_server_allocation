```python
import pandas as pd
import ast
```

```python
topology = 'germany'
model = 'm2'

solution = 'heuristic'
# solution = 'linear'
```

```python

df = pd.read_csv(topology+'/'+model+'/csv/heuristic_1.csv', header=None)
```

```python
def get_data(data):
    # Use ast.literal_eval to safely evaluate the string
    data = ast.literal_eval(data)

    # Extract the components
    list_of_lists = data[0]
    list_of_strings = data[1]

    return list_of_lists, list_of_strings
```

```python
# for each df
dict_placement = dict()
dict_placement_1 = dict()
for i in range(len(df)):
    # print(df.iloc[i][1])
    placement, time = get_data(df.iloc[i][1])

    print("List of Lists:", placement)
    print("List of Strings:", time)

    dict_placement[df.iloc[i][0]] = time
    dict_placement_1[df.iloc[i][0]] = placement


dict_sorted = dict()

# print order by key
for i in range(0, 20):
    app = 'App' + str(i)
    # print("%s: %s" % (key, dict_placement[key]))
    dict_sorted[app] = dict_placement[app]
```
